```python
'''Image Blurring refers to making the image less clear or distinct. 
It is done with the help of various low pass filter kernels'''
```

<p style="font-family:consolas; font-size: 24px; color: cyan"> Advantages of blurring:</p>

```python
'''
#It helps in Noise removal. As noise is considered as high pass signal so by the application of low pass filter kernel 
# we restrict noise.
#It helps in smoothing the image.
#Low intensity edges are removed.
#It helps in hiding the details when necessary. For e.g. in many cases police deliberately want 
# to hide the face of the victim, in such cases blurring is required.
'''
```

<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Important types of blurring:</p>

```python
'''Gaussian Blurring:Gaussian blur is the result of blurring an image by a Gaussian function.'''
'''Median Blur: The Median Filter is a non-linear digital filtering technique, 
often used to remove noise from an image or signal. '''
'''Bilateral Blur: A bilateral filter is a non-linear, edge-preserving, and noise-reducing smoothing filter for images'''
```

<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Load image</p>

```python
# importing libraries
import cv2
import numpy as np

from matplotlib import pyplot as plt
# Load the image


img = cv2.imread('./data/pexels-saab-340.jpg')
 
# Add the first image to the figure (top-left position)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, first position

# Gaussian Blur
Gaussian = cv2.GaussianBlur(img, (7, 7), 0)
plt.imshow(cv2.cvtColor(Gaussian,cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis labels
plt.title("Gaussian Blurring")  # Optional: Add a title


# Add the second image to the figure (top-right position)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, second position
# Median Blur
median  = cv2.medianBlur(img, 5)
plt.imshow(cv2.cvtColor(median ,cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis labels
plt.title("Median Blurring")  # Optional: Add a title

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, third position
## Bilateral Blur
bilateral  = cv2.bilateralFilter(img, 9, 75, 75)
plt.imshow(cv2.cvtColor(bilateral,cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis labels
plt.title("Bilateral Blur")  # Optional: Add a title

# Add the fourth image to the figure (bottom-right position)
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, fourth position

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis labels
plt.title("Initial Image") # Optional: Add a title

plt.show()
```

```python
def make_more_vivid(image):
    """Modify the saturation and value of the image."""
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hue, saturation, value = cv2.split(hsv)

    saturation = np.array(saturation * 1.2, dtype=np.uint16)
    saturation = np.array(np.clip(saturation, 0, 255), dtype=np.uint8)

    value = np.array(value * 1.1, dtype=np.uint16)
    value = np.array(np.clip(value, 0, 255), dtype=np.uint8)

    return cv2.cvtColor(cv2.merge((hue, saturation, value)), cv2.COLOR_HSV2BGR)
```

```python
def process(image):
    """Given an image process it using the process to replicate a lens blur."""
    
    image = make_more_vivid(image)
    image = cv2.bilateralFilter(image, 9, 150, 150)
    image = cv2.blur(image, (15, 15))

    return image
```

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position

image1 = process(img)
plt.imshow(cv2.cvtColor(image1 ,cv2.COLOR_BGR2RGB))
 
plt.axis('off')  # Hide the axis labels
plt.title("replicate a lens blur")  # Optional: Add a title


# Add the second image to the figure (top-right position)
plt.subplot(1, 2, 2)  # 1 rows, 2 columns, second position

image2 = make_more_vivid(img)
plt.imshow(cv2.cvtColor(image2 ,cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis labels
plt.title("make_more_vivid(image)")  # Optional: Add a title
 
plt.show()
```
