<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Adjusting brightness and contrast</p>

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> One common method is to use the cv2.addWeighted() function</p>

```python
#Import the necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Adjust the brightness and contrast
# Adjusts the brightness by adding 10 to each pixel value
brightness = 10 
# Adjusts the contrast by scaling the pixel values by 2.3
contrast = 2.3  
img = cv2.addWeighted(img, contrast, np.zeros(img.shape, img.dtype), 0, brightness)

#Save the image
cv2.imwrite('./data/add-modified_pexels-saab-340.jpg', img)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Brightness & contrast")
plt.imshow(img)
```

```python
'''In this example, the brightness of the image is adjusted by adding 10 to each pixel value, 
and the contrast is adjusted by scaling the pixel values by 2.3'''
```

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> Another method is to use the cv2.convertScaleAbs() function</p>

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Adjust the brightness and contrast 
# g(i,j)=α⋅f(i,j)+β
# control Contrast by 1.5
alpha = 1.5  
# control brightness by 50
beta = 50  
img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

#Save the image
cv2.imwrite('./data/abs-modified_pexels-saab-340.jpg', img)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Brightness & contrast")
plt.imshow(img)
```

```python
'''In this example, the brightness and contrast are adjusted using a combination of scaling and shifting the pixel values'''
```

<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Sharpening images</p>

```python
'''Sharpening is the process of enhancing the edges and fine details in an image to make it appear sharper 
and more defined. It is important because it can help to bring out the details and features in an image, 
making it more visually appealing and easier to understand'''
```

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> One common method is to use the cv2.filter2D() function, which convolves the image with a kernel</p>

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Create the sharpening kernel
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

# Sharpen the image
sharpened_image = cv2.filter2D(img, -1, kernel)

#Save the image
cv2.imwrite('./data/sharpened-modified_pexels-saab-340.jpg', sharpened_image)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Sharpening")
plt.imshow(sharpened_image)
```

```python
'''In this example, a 3x3 sharpening kernel is used to enhance the edges in the image. 
You can experiment with different kernels to achieve the desired level of sharpening'''
```

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> Another method for sharpening images is to use the cv2.Laplacian() function</p>

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Create the sharpening kernel
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

# Sharpen the image using the Laplacian operator
sharpened_image2 = cv2.Laplacian(img, cv2.CV_64F)

#Save the image
cv2.imwrite('./data/laplacian-modified_pexels-saab-340.jpg', sharpened_image2)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Laplacian Sharpening")
plt.imshow(sharpened_image2)
```

<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Removing noise from images</p>

```python
'''Noise reduction is the process of removing or reducing unwanted noise or artifacts from an image. '''
```

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> One common method for removing noise is to use a median filter: cv2.medianBlur() function</p>

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Remove noise using a median filter
filtered_image = cv2.medianBlur(img, 5)

#Save the image
cv2.imwrite('./data/median-modified_pexels-saab-340.jpg', filtered_image)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Median Blur")
plt.imshow(filtered_image)
```

```python
'''In this example, the cv2.medianBlur() function is used to apply a median filter to the image. 
The 5 parameter specifies the size of the kernel to use for the filter'''
```

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> Another common method for removing noise is to use a Gaussian filter: cv2.GaussianBlur() function</p>

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Remove noise using a Gaussian filter
filtered_image2 = cv2.GaussianBlur(img, (7, 7), 0)

#Save the image
cv2.imwrite('./data/gaussian-modified_pexels-saab-340.jpg', filtered_image2)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Gaussian Blur")
plt.imshow(filtered_image2)
```

<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Enhancing color in images</p>

```python
'''Color enhancement is adjusting the colors in an image to make them more vibrant, balanced, or natural. 
It can be used to correct color defects or problems in an image or
to simply make an image more appealing and aesthetically pleasing'''
```

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> One common method is to use the cv2.cvtColor() function</p>


```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)


# Convert the image from BGR to HSV color space
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

# Adjust the hue, saturation, and value of the image
# Adjusts the hue by multiplying it by 0.7
img[:, :, 0] = img[:, :, 0] * 0.7
# Adjusts the saturation by multiplying it by 1.5
img[:, :, 1] = img[:, :, 1] * 1.5
# Adjusts the value by multiplying it by 0.5
img[:, :, 2] = img[:, :, 2] * 0.5

# Convert the image back to BGR color space
image2 = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)

#Save the image
cv2.imwrite('./data/coloured-modified_pexels-saab-340.jpg', image2)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Enhanced coloured")
plt.imshow(image2)
```

```python
'''This code first converts the image from the BGR color space to the HSV color space using the cv2.cvtColor() function. 
It then adjusts the hue, saturation, and value (brightness) of the image 
by multiplying the corresponding channels by a scalar value'''
```

<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Image resizing and scaling</p>

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> One common method is to use the cv2.resize() function</p>

```python
'''Image resizing and scaling involve adjusting the dimensions and size of an image.'''
'''
cv2.resize() function takes the following arguments:
    src: The input image.
    dsize: The size of the output image, specified as a tuple (width, height).
    fx: The scaling factor along the x-axis.
    fy: The scaling factor along the y-axis.
    interpolation: The interpolation method to use.
'''
```

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Resize the image to a specific width and height
resized_image = cv2.resize(img, (300, 200))

#Save the image
cv2.imwrite('./data/resized-modified_pexels-saab-340.jpg', resized_image)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Resized")
plt.imshow(resized_image)
```

```python
'''In this example, the image is resized to a width of 300 pixels and a height of 200 pixels.'''
```

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Scale the image by a factor of 2 along both axes
scaled_image = cv2.resize(img, None, fx=2.5, fy=2)

#Save the image
cv2.imwrite('./data/scaled-modified_pexels-saab-340.jpg', scaled_image)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Scaled")
plt.imshow(scaled_image)
```

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Inverse by subtracting from 255
inverse_image = 255 - img

#Save the image
cv2.imwrite('./data/inverse-modified_pexels-saab-340.jpg', inverse_image )
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Inverse color")
plt.imshow(inverse_image )
```

<p style="font-family:Roboto; font-size: 22px; color: orange; text-decoration-line: overline; "> Part: _Equalizing histograms</p>

```python
'''Histogram equalization is a technique used to adjust the contrast of an image by spreading out 
the intensity values of the pixels in the image'''
```

<p style="font-family:Roboto; font-size: 22px; color: lightgreen; text-decoration-line: overline; "> One common method is to use the cv2.equalizeHist() function</p>

```python
img = cv2.imread('./data/pexels-saab-340.jpg')

# Add the first image to the figure (top-left position)
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, first position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Ooriginal image")
# Adjust layout to prevent overlap
plt.tight_layout()
plt.imshow(img)

# Convert the image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Equalize the histogram
equalized_image = cv2.equalizeHist(gray_image)

#Save the image
cv2.imwrite('./data/equalized-modified_pexels-saab-340.jpg', equalized_image)
#Plot the contrast image
plt.subplot(1, 2, 2) # 1 rows, 2 columns, second position
plt.axis('off')  # Optional: Turn off axes for a cleaner look
plt.title("Equalized color")
plt.imshow(equalized_image )
```
